# Netflix Data Quality Enhancement and Preprocessing using Python

## Objective
The objective of this project is to clean and preprocess the Netflix Movies and TV Shows dataset by handling missing values, checking duplicates, standardizing formats, validating data quality, and preparing the dataset for analysis.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("netflix_titles.csv")

print("Dataset Loaded Successfully")
```

# Dataset Overview

This section provides an overview of the dataset structure, including the number of rows, columns, and sample records.

```python
print("Dataset Shape:", df.shape)

print("\nColumns:")
print(df.columns)

df.head()
```

# Dataset Information

This section provides detailed information about the dataset, including the number of entries, column names, data types, and non-null values. It helps in understanding the structure of the dataset and identifying potential data quality issues.

```python
df.info()
```

## Observation

This section helps identify data types and missing values in each column.

# Missing Value Analysis

This section identifies missing values present in the dataset and helps determine the appropriate data cleaning strategy.

```python
missing_values = df.isnull().sum()

print(missing_values)
```

## Observation

The dataset contains missing values in columns such as director, cast, country, date_added, and rating. These missing values need to be handled before further analysis.

# Duplicate Record Analysis

This section checks whether the dataset contains any duplicate records that may affect data quality and analysis results.

```python
duplicates = df.duplicated().sum()

print("Duplicate Records:", duplicates)
```

## Observation

The dataset was checked for duplicate records. Any duplicate entries identified may need to be removed to maintain data accuracy and consistency.

# Column Name Standardization

This section standardizes column names by converting them to lowercase and replacing spaces with underscores for better readability and consistency.

```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)

print(df.columns)
```

## Observation

Column names have been standardized to ensure consistency and make them easier to reference during analysis.

# Missing Value Treatment

This section handles missing values present in the dataset using appropriate replacement techniques.

```python
df["director"] = df["director"].fillna("Unknown")

df["cast"] = df["cast"].fillna("Not Available")

df["country"] = df["country"].fillna("Unknown")

df["rating"] = df["rating"].fillna("Not Rated")

df["date_added"] = df["date_added"].fillna("Unknown")
```

## Observation

Missing values were replaced with meaningful placeholders to preserve records and improve data completeness.

# Text Standardization

This section standardizes text values to ensure consistency across categorical variables.

```python
df["type"] = df["type"].str.title()

df["country"] = df["country"].str.title()
```

## Observation

Categorical text values were standardized to maintain uniform formatting throughout the dataset.

# Date Format Standardization

This section converts date values into a consistent datetime format for easier analysis and processing.

```python
df["date_added"] = pd.to_datetime(
    df["date_added"],
    errors="coerce"
)

print(df["date_added"].head())
```

## Observation

Date values were successfully converted into a standard datetime format, improving consistency and usability.

```python
df["date_added"] = df["date_added"].fillna(pd.Timestamp("2000-01-01"))

df["duration"] = df["duration"].fillna("Unknown")

print(df.isnull().sum())
```

# Data Type Verification

This section verifies the data types of all columns after preprocessing.

print(df.dtypes)

## Observation

The dataset was reviewed to ensure that each column has an appropriate data type for analysis.

# Data Validation Checks

This section performs validation checks to identify records containing invalid or unusual values.

```python
invalid_years = df[
    (df["release_year"] < 1900) |
    (df["release_year"] > 2025)
]

print("Invalid Release Years:", len(invalid_years))
```

## Observation

Release year values were validated to identify records that fall outside the expected range.

# Final Missing Value Verification
This section verifies whether missing values have been successfully handled after preprocessing.


```python
print(df.isnull().sum())
```

## Observation

A final review was conducted to confirm that missing values were appropriately addressed.

# Data Quality Report

This section summarizes the overall data cleaning and preprocessing activities performed on the dataset.

```python
report = f"""
NETFLIX DATA QUALITY REPORT

Total Records: {df.shape[0]}
Total Columns: {df.shape[1]}

Missing Values Remaining:
{df.isnull().sum().sum()}

Duplicate Records:
{duplicates}

Invalid Release Years:
{len(invalid_years)}
"""

print(report)
```

## Observation

The report provides a summary of the dataset quality after cleaning and preprocessing.

# Export Cleaning Report

This section saves the generated data quality report as a text file.

```python
with open("cleaning_report.txt", "w") as file:
    file.write(report)

print("Report Saved Successfully")
```

## Observation

The data quality report was successfully exported for documentation purposes.

# Export Clean Dataset

This section saves the cleaned dataset for future analysis and visualization.

```python
df.to_csv(
    "cleaned_netflix_titles.csv",
    index=False
)

print("Clean Dataset Saved Successfully")
```

The cleaned dataset was successfully exported and is ready for further analysis.
